## data cleaning and analysis

```python
df4 = pd.DataFrame({"name" : ['Alfred','Batman','Catwoman'],
                    "toy" : [np.nan, 'Batmobile', 'Bullwhip'],
                    "DOM" : [pd.NaT, pd.Timestamp("1980-04-25"), pd.NaT]})

df4.head()
```

```python
df4.dropna() # deletes rows with any null value
```

```python
df4 = pd.DataFrame({"name" : ['Alfred','Batman','Catwoman'],
                    "toy" : [np.nan, np.nan, np.nan],
                    "DOM" : [pd.NaT, pd.Timestamp("1980-04-25"), pd.NaT]})

df4
```

```python
df4.dropna(how='all', axis=1) #deletes columns with all null values
```

```python
df4 = pd.DataFrame({"name" : ['Alfred','Batman','Alfred'],
                    "toy" : [np.nan, np.nan, np.nan],
                    "DOM" : [pd.NaT, pd.Timestamp("1980-04-25"), pd.NaT]})
df4
```

```python
df4.drop_duplicates(subset="name")  #subset specifies which columns to check for duplicates
```

```python
# to count no. of unique values in any column
df4['toy'].value_counts(dropna = False)
```
